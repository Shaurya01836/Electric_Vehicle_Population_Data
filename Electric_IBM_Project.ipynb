```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload the file
uploaded = files.upload()

# Load the dataset
file_path = 'Electric_Vehicle_Population_Data.csv'  # The file will be uploaded to the current directory
df = pd.read_csv(file_path)

# Basic Data Exploration
def data_exploration(df):
    print("First 5 rows of the dataset:")
    display(df.head())

    print("\nDataset Information:")
    df.info()

    print("\nSummary Statistics:")
    display(df.describe())

    print("\nChecking for missing values:")
    display(df.isnull().sum())

data_exploration(df)

# Data Cleaning
def data_cleaning(df):
    # Handle missing values
    df = df.dropna()  # Dropping rows with missing values, adjust as necessary

    # Convert columns to appropriate data types if necessary
    # Example: df['column_name'] = df['column_name'].astype('int')

    print("Data after cleaning:")
    display(df.head())

    return df

df = data_cleaning(df)

# Data Visualization
def data_visualization(df):
    # Electric Vehicles by Make
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='Make', order=df['Make'].value_counts().index)
    plt.xticks(rotation=90)
    plt.title('Number of Electric Vehicles by Make')
    plt.show()

    # Electric Vehicles by Model Year
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='Model Year', order=df['Model Year'].value_counts().index)
    plt.title('Number of Electric Vehicles by Model Year')
    plt.show()

    # Electric Vehicles by State
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='State', order=df['State'].value_counts().index)
    plt.xticks(rotation=90)
    plt.title('Number of Electric Vehicles by State')
    plt.show()



    # Trend of Electric Vehicle Registration Over Years
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df, x='Model Year', y=df.groupby('Model Year').size(), marker='o')
    plt.title('Trend of Electric Vehicle Registration Over Years')
    plt.show()

    # Distribution of Electric Range
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Electric Range'], bins=30, kde=True)
    plt.title('Distribution of Electric Range')
    plt.xlabel('Electric Range')
    plt.ylabel('Frequency')
    plt.show()

    # Scatter Plot of Electric Range vs. Model Year
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='Model Year', y='Electric Range')
    plt.title('Scatter Plot of Electric Range vs. Model Year')
    plt.xlabel('Model Year')
    plt.ylabel('Electric Range')
    plt.show()

data_visualization(df)

# Basic Data Analysis
def data_analysis(df):
    make_counts = df['Make'].value_counts()
    print("Electric Vehicle counts by Make:")
    display(make_counts)

    model_year_counts = df['Model Year'].value_counts()
    print("Electric Vehicle counts by Model Year:")
    display(model_year_counts)

    top_models = df['Model'].value_counts().head(10)
    print("Top 10 Electric Vehicle Models:")
    display(top_models)

data_analysis(df)
```
